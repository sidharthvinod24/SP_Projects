# Dataset(1) Visitor International Arrivals To Singapore By Country Monthly

```python
#Sidharth Vinod
#P1936252
#DIT/FT/1B/11

import pandas as pd
import matplotlib.pyplot as plt
#This is the dataset that needs to be loaded

df=pd.read_csv("dataset_PDS/visitor-international-arrivals-to-singapore-by-country-monthly.csv")
print("Successfully loaded dataset dataset_PDS/visitor-international-arrivals-to-singapore-by-country-monthly.csv")
print()
#This is the shape of the dataset
print("This is the shape of the dataset \n{}".format(df.shape))
print()
#This is the index of the dataset
print("This is the index of the dataset \n{}".format(df.index))
print()

#Columns of the dataset
print("This are the columns in the dataset \n{}".format(df.columns))

#No of non-NA values 
print("The total number of non-NA values in this dataset is: \n{}".format(df.count))

#A summmary of the dataset
print("A summary of this dataset is shown below \n{}".format(df.info))

#A descriptive statistical summary of no of visitor arrivals
print("A descriptive statistical summary of this dataset is shown below: \n{}".format(df['no_of_visitor_arrivals'].describe()))
```

```python
#Lets Turn Months into years for Ease of analysis
df['month']=pd.to_datetime(df['month'])
df['year']=df.month.dt.year
df
```

```python
#Lets Create a Bar Chart to find out which region has the highest international arrival on average
results=df.groupby('region').mean()
results
regions=[region for region, df in df.groupby('region')]
regions
x=df[df['region']=='Europe']['no_of_visitor_arrivals']
print(x.mean())
plt.figure(1,figsize=(5,5))
barchart=plt.bar(regions,results['no_of_visitor_arrivals'])
plt.xticks(regions,size=8,fontsize=15)
plt.ylabel('Average No of International Visitors',fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.title("Average Number Of International Arrivals By Region")
plt.show()
```

```python
#Lets Create A Line Chart to show the trend of the International Arrivals from (1978-2015)
df1=df.pivot_table('no_of_visitor_arrivals','year','region')
plt.figure(1,figsize=(5,5))
plt.title("No of Visitor Arrivals From Region")
plt.plot(df1['Asia'],'b',label='Asia')
plt.plot(df1['Americas'],'r',label='Americas')
plt.plot(df1['Europe'],'g',label='Europe')
plt.plot(df1['Oceania'],'m',label='Oceania')

plt.legend(loc='upper left');
```

```python
#Lets Create a Piechart comparing the International Arrivals in 1978 and 2015
fig=plt.figure(figsize=(15,15))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
pd.DataFrame(df)
year=df[df['year']==1978]
regions=[region for region, year in year.groupby('region')]
results=year.groupby('region').sum()
labels = regions
values = results['no_of_visitor_arrivals']
explode = [0,0.1,0,0]
ax1.pie(values,labels=labels,explode=explode,shadow=True,startangle=90,autopct='%1.1f%%')
ax1.set_title("Number Of International Arrivals in 1978")

#Piechart for the latest year which is 2015
year=df[df['year']==2015]
regions=[region for region, year in year.groupby('region')]
results=year.groupby('region').sum()
labels = regions
values = results['no_of_visitor_arrivals']
explode = [0,0.1,0,0]
ax2.pie(values,labels=labels,explode=explode,shadow=True,startangle=90,autopct='%1.1f%%')
ax2.set_title("Number Of International Arrivals in 2015")
plt.show()
```

```python
#Lets create a Scatter Plot to see whether there is a relationship
#between Arrivals from Oceania and from Americas
import numpy as np
x=df[df['region']=='Oceania']['no_of_visitor_arrivals']
y=df[df['region']=='Americas']['no_of_visitor_arrivals']
plt.figure(figsize=(7,6))
plt.scatter(x,y,linewidth=2,alpha=0.75)
plt.title("Scatter Plot of International Arrivals From Oceania and Americas")
plt.xlabel("International Arrivals From Oceania")
plt.ylabel("International Arrivals From Americas")
#I used numpy to create a line of best fit
z=np.polyfit(x,y,1)
p=np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show()
```

# Dataset(2) Monthly Gazetted Hotel Statistics By Tier

```python
#Dataset 2
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('dataset_PDS/monthly-gazetted-hotel-statistics-by-tier.csv')
#This is the shape of the dataset
print("This is the shape of the dataset \n{}".format(df.shape))
print()
#This is the index of the dataset
print("This is the index of the dataset \n{}".format(df.index))
print()

#Columns of the dataset
print("This are the columns in the dataset \n{}".format(df.columns))
print()

#No of non-NA values 
print("The total number of non-NA values in this dataset is: \n{}".format(df.count))
print()

#A summmary of the dataset
print("A summary of this dataset is shown below ")
print(df.info())
print()

#A descriptive statistical summary of Standard Average Occupancy Rate
print("A descriptive statistical summary of Standard Average Occupancy Rate(%) is shown below: \n{}".format(df['aor'].describe()))
print()

#A descriptive statistical summary of Standard Average Room Rate
print("A descriptive statistical summary of Standard Average Room Rate is shown below: \n{}".format(df['arr'].describe()))
print()

#A descriptive statistical summary of Revenue Per Available Room
print("A descriptive statistical summary of Revenue Per Available Room is shown below: \n{}".format(df['revpar'].describe()))
print()

```

```python
# Lets turn the months into years
df['period']=pd.to_datetime(df['period'])
df['year']=df.period.dt.year
df
```

```python
#Lets create a line chart to show the trend of 
#Standard Average Occupancy Rate by Hotel Tier
df1=df.pivot_table('aor','period','tier')
df1
plt.figure(1,figsize=(7,7))
plt.title("Standard Average Occupancy Rate By Hotel Tier")
plt.plot(df1['ECONOMY'],'b',label='Economy')
plt.plot(df1['LUXURY'],'r',label='Luxury')
plt.plot(df1['MID-TIER'],'g',label='Mid-Tier')
plt.plot(df1['UPSCALE'],'m',label='Upscale')
plt.legend(loc='upper left');
```

```python
#Lets Create a Bar Chart to find out which 
#Hotel Tier has the highest Standard Average Room Rate
results=df.groupby('tier').mean()
results
tiers=[tier for tier, df in df.groupby('tier')]
tiers
plt.figure(1,figsize=(6,5))
plt.bar(tiers,results['arr'])
plt.xticks(tiers,size=8,fontsize=15)
plt.ylabel('Standard Average Room Rate',fontsize=15)
plt.xlabel('Hotel Tier',fontsize=15)
plt.title("Standard Average Room Rate by Hotel Tier")
plt.show()
```

```python
#Lets see which hotel tier has the highest revenue per room
tier=[tier for tier, df in df.groupby('tier')]
results=df.groupby('tier').sum()
labels = tier
values = results['revpar']
explode = [0,0.1,0,0]
plt.figure(1,figsize=(5,5))
plt.pie(values,labels=labels,explode=explode,shadow=True,startangle=90,
autopct='%1.1f%%')
plt.title("Percentage of Revenue Per Avalilable Room For Different Tiers")
plt.show()
```

# Dataset 3( Total Air Cargo Discharged In Tonnes By Country)

```python
#Dataset 3
import pandas as pd
import matplotlib.pyplot as plt
#This are the missing values in the dataset 
missing_values=['-','na']
df=pd.read_csv('dataset_PDS/total-air-cargo-discharged-in-tonnes-by-country.csv',na_values=missing_values)
#This is the shape of the dataset
print("This is the shape of the dataset \n{}".format(df.shape))
print()
#This is the index of the dataset
print("This is the index of the dataset \n{}".format(df.index))
print()

#Columns of the dataset
print("This are the columns in the dataset \n{}".format(df.columns))
print()

#No of non-NA values 
print("The total number of non-NA values in this dataset is: \n{}".format(df.count))
print()

#A summmary of the dataset
print("A summary of this dataset is shown below ")
print(df.info())
print()

#A descriptive statistical summary of no of visitor departure
print("A descriptive statistical summary of this dataset is shown below: \n{}".format(df['value'].describe()))
print()
```

```python
#Lets turn the months into years
df['month']=pd.to_datetime(df['month'])
df['year']=df.month.dt.year
#To replace the NaN values with zero we use df.fillna(0) method
df.value=df.value.fillna(0)
#We drop the level_1 column as it contains the same information
df.drop('level_1' ,inplace=True,axis=1)
df
```

```python
#Lets create a line chart to show the trend of 
#Air Cargo to other region over the years
df1=df.pivot_table('value','year','level_2')
df1
plt.figure(1,figsize=(5,5))
plt.title("Amount Of Air Cargo Discharged In Tonnes From Singapore To Other Regions")
plt.plot(df1['Europe'],'b--',label='Europe')
plt.plot(df1['North East Asia'],'r--',label='North East Asia')
plt.plot(df1['South East Asia'],'g--',label='South East Asia')
plt.legend(loc='upper left');
```

```python
#Lets create a box plot to show the distribution 
#of Air Cargo within each region
fig=plt.figure(figsize=(5,5))

north=df[df.level_2=='North East Asia']['value']
south=df[df.level_2=='South East Asia']['value']
europe=df[df.level_2=='Europe']['value']

data_number=[north,south,europe]
bp=plt.boxplot(data_number,labels=['North East Asia','South East Asia',
'Europe'],patch_artist=True)
for patch in bp["boxes"]:
    patch.set(facecolor="cyan")
plt.title("Boxplot for Air Cargo Discharged By Singapore")
plt.xlabel("Regions")
plt.ylabel("Amount Of Air Cargo Discharged In Tonnes")
plt.show()
```
